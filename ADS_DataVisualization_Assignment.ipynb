```python
import plotly.express as px
df = px.data.gapminder()
df.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (14, 8))
plt.scatter(df["gdpPercap"], df["lifeExp"], 
            s=df["pop"]*0.00001, alpha=0.5)
plt.show()
```

```python
df[["lifeExp", "continent"]].mean()
```
